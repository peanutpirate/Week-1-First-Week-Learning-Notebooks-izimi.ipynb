```python
# Pandas ile Temel Grafik Ã‡izimi (Plotly/Seaborn)

## AmaÃ§: Bu notebookta Pandas kullanarak aÅŸaÄŸÄ±daki grafikleri oluÅŸturacaÄŸÄ±m.
# 1. Ã‡izgi Grafikleri (Line Plots): ðŸ“ˆ Verilerin zaman iÃ§indeki deÄŸiÅŸimini gÃ¶rmek iÃ§in idealdir.
# 2. SÃ¼tun Grafikleri (Bar Plots): ðŸ“Š FarklÄ± kategorileri karÅŸÄ±laÅŸtÄ±rmak iÃ§in kullanÄ±lÄ±r. (Ã–rneÄŸin: Åžehirlere gÃ¶re satÄ±ÅŸ miktarlarÄ±).
# 3. Histogramlar: ðŸ§± Verilerin daÄŸÄ±lÄ±mÄ±nÄ± anlamamÄ±za yardÄ±mcÄ± olur. (Ã–rneÄŸin: Bir gruptaki insanlarÄ±n yaÅŸ aralÄ±klarÄ±).


## Analiz Ã–ncesi DÃ¼ÅŸÃ¼nceler:
# 1. Bu veri hangi soruya cevap verebilir?
# 2. Ä°lk bakmak istediÄŸim 3 ÅŸey:
# - Hangi grafik gÃ¶ze daha hoÅŸ gÃ¶rÃ¼nÃ¼yor.
# - Hangi kÃ¼tÃ¼phaneyi Ã§aÄŸrÄ±nca hangi grafik geliyor. ( import matplotlib.pyplot as plt ) (GrafiÄŸi ekranda gÃ¶stermek iÃ§in gerekli!)
# - BaÅŸka grafik olarak dataya ne ekleyebiliyoruz.
# 3. Veriyle ilgili varsayÄ±mÄ±m: Bence gÃ¶rselleÅŸtirmek datayÄ± anlaÅŸÄ±labilir kÄ±lÄ±yor.

import pandas as pd
import matplotlib.pyplot as plt

# Ã–rnek veri: Pandas Ã¶ÄŸrenme gÃ¼nlerin ve baÅŸarÄ± puanÄ±n
veri = {
    'Gun': [1, 2, 3, 4, 5, 6, 7],
    'Basari_Puani': [10, 25, 20, 45, 50, 75, 90]
}

df = pd.DataFrame(veri)

# GrafiÄŸi Ã§izdirelim
df.plot(x='Gun', y='Basari_Puani', kind='line', marker='o', color='teal')

# BaÅŸlÄ±k ve etiketler ekleyelim (Opsiyonel ama iyi bir pratik)
plt.title('Pandas Ã–ÄŸrenme YolculuÄŸum')
plt.xlabel('GÃ¼nler')
plt.ylabel('BaÅŸarÄ± PuanÄ±')

# GrafiÄŸi gÃ¶ster:
plt.show()

```

```python
## Grafik Yorumu (Ã‡izgi GrafiÄŸi/Line Plots)
# - Grafik neyi netleÅŸtirdi?
# --Neden Bu GrafiÄŸi KullanÄ±rÄ±z? (Line grafik)
# -- Trend Analizi: Verinin artÄ±ÅŸ mÄ± yoksa azalÄ±ÅŸ mÄ± gÃ¶sterdiÄŸini anÄ±nda anlarsÄ±n.
# -- PÃ¼f NoktasÄ±: marker='o' parametresi, her bir veri noktasÄ±nÄ±n Ã¼zerine kÃ¼Ã§Ã¼k birer yuvarlak koyar.
# ---Bu, verinin tam olarak hangi noktada olduÄŸunu gÃ¶rmeni saÄŸlar.
# - SayÄ±yla fark etmediÄŸim neyi fark ettim?
# # -- KÄ±rÄ±lma NoktalarÄ±: Ã–rneÄŸin 3. gÃ¼nde baÅŸarÄ± puanÄ±nda kÃ¼Ã§Ã¼k bir dÃ¼ÅŸÃ¼ÅŸ olduÄŸunu (25'ten 20'ye) grafikteki "v" ÅŸeklinden hemen fark edebilirsin.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Yeni veri setimiz
veriler = {
    'Konu': ['Seriler', 'DataFrame', 'Filtreleme', 'GÃ¶rselleÅŸtirme', 'Eksik Veri'],
    'Saat': [4, 8, 5, 3, 6]
}

df_bar = pd.DataFrame(veriler)

# SÃ¼tun grafikini Ã§izdirelim (kind='bar')
df_bar.plot(x='Konu', y='Saat', kind='bar', color='orange', edgecolor='black')

# Grafik dÃ¼zenlemeleri
plt.title('Pandas KonularÄ±na GÃ¶re Ã‡alÄ±ÅŸma SÃ¼releri')
plt.ylabel('Harcanan Saat')
plt.xticks(rotation=45) # YazÄ±lar birbirine girmesin diye 45 derece dÃ¶ndÃ¼rdÃ¼k
plt.show()
```

```python
## Grafik Yorumu (Ã‡izgi GrafiÄŸi/Line Plots)
# - Grafik neyi netleÅŸtirdi?
# --Neden Bu GrafiÄŸi KullanÄ±rÄ±z? (SÃ¼tun Grafikleri (Bar Plots))
# -- Kategorik Veri: SÃ¼tun grafikleri genellikle metin tabanlÄ± (Seriler, DataFrame vb.) kategoriler iÃ§in en iyisidir.
# -- SÄ±ralama: Verini bÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralarsan grafiÄŸin Ã§ok daha profesyonel gÃ¶rÃ¼nÃ¼r.
# ---(Bunun iÃ§in Pandas'Ä±n .sort_values() fonksiyonunu kullanabilirsin).
# --- EÄŸer sÃ¼tunlarÄ±n yatay olmasÄ±nÄ± istersen kind='barh' (bar horizontal) yazman yeterli.
## - Bu, Ã¶zellikle kategori isimleri Ã§ok uzun olduÄŸunda hayat kurtarÄ±r.
```

```python
## Histogramlar
# SÃ¼tun grafikleri (Bar Plots) kategorileri karÅŸÄ±laÅŸtÄ±rÄ±rken, Histogramlar sayÄ±sal verilerin nasÄ±l daÄŸÄ±ldÄ±ÄŸÄ±nÄ± anlamamÄ±zÄ± saÄŸlar.
# Ã–rneÄŸin; "Verilerim hangi aralÄ±kta yoÄŸunlaÅŸÄ±yor?", "UÃ§ deÄŸerler (aykÄ±rÄ± veriler) var mÄ±?" gibi sorularÄ±n cevabÄ±nÄ± burada buluruz.

## Senaryo: Ã–ÄŸrenci Not DaÄŸÄ±lÄ±mÄ±
# -Diyelim ki elinde 100 kiÅŸilik bir sÄ±nÄ±fÄ±n Pandas sÄ±nav notlarÄ± var.
# -Kimin kaÃ§ aldÄ±ÄŸÄ±ndan ziyade, sÄ±nÄ±fÄ±n genel baÅŸarÄ±sÄ±nÄ± (Ã§oÄŸunluk 50 mi aldÄ± yoksa 90 mÄ±?) gÃ¶rmek istiyorsun.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Rastgele 100 tane sÄ±nav notu oluÅŸturalÄ±m (0-100 arasÄ±)
notlar = {'Sinav_Notu': [45, 55, 60, 65, 70, 72, 75, 80, 82, 85, 88, 90, 92, 40, 50, 55]}
df_hist = pd.DataFrame(notlar)

# Histogram Ã§izimi (kind='hist')
# 'bins' parametresi veriyi kaÃ§ kutucuÄŸa bÃ¶leceÄŸimizi belirler
df_hist['Sinav_Notu'].plot(kind='hist', bins=5, color='purple', edgecolor='white')

plt.title('SÄ±nav NotlarÄ±nÄ±n DaÄŸÄ±lÄ±mÄ±')
plt.xlabel('Not AralÄ±ÄŸÄ±')
plt.ylabel('Ã–ÄŸrenci SayÄ±sÄ±')
plt.show()
```

```python
## Grafik Yorumu (Histogramlar)
# - Grafik neyi netleÅŸtirdi?
# --Neden Bu GrafiÄŸi KullanÄ±rÄ±z? (Histogramlar)
# -- HistogramÄ±n GÃ¼cÃ¼: bins Parametresi. Histogramda en Ã¶nemli oyuncu bins (kutucuklar) deÄŸeridir:
# -- KÃ¼Ã§Ã¼k bins: Veriyi Ã§ok geniÅŸ aralÄ±klarla gruplar (detayÄ± kaybedersin).
# -- BÃ¼yÃ¼k bins: Veriyi Ã§ok kÃ¼Ã§Ã¼k parÃ§alara bÃ¶ler (grafik Ã§ok "tÄ±rtÄ±klÄ±" gÃ¶rÃ¼nÃ¼r).
# -- Ä°deal olan: Verinin genel ÅŸeklini (Ã§an eÄŸrisi mi, saÄŸa yatÄ±k mÄ±?) anlayabildiÄŸin bir deÄŸer seÃ§mektir.
```

```python
## Åžimdi bu temel grafiklerin (Ã‡izgi, SÃ¼tun, Histogram) hepsini tek bir gÃ¶rselde toplamayÄ± gÃ¶relim: "Subplots" (Alt Grafikler)
# Pandas ve Matplotlib kullanarak ekranÄ± parÃ§alara bÃ¶lebilir ve her bir bÃ¶lmeye farklÄ± bir grafik yerleÅŸtirebiliriz.
import pandas as pd
import matplotlib.pyplot as plt

# 1. Verimizi hazÄ±rlayalÄ±m
veri = {
    'Gun': [1, 2, 3, 4, 5],
    'Satis': [10, 20, 15, 25, 30],
    'Kategori': ['A', 'B', 'C', 'D', 'E'],
    'Deger': [5, 15, 10, 20, 25],
    'Dagilim': [10, 12, 12, 13, 15, 20, 22, 25, 30, 35] # Histogram iÃ§in
}

# 2. FigÃ¼r ve Eksenleri (Axes) oluÅŸturalÄ±m
# 1 satÄ±r, 3 sÃ¼tunluk bir alan yaratÄ±yoruz
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# --- 1. Grafik: Ã‡izgi GrafiÄŸi ---
ax[0].plot(veri['Gun'], veri['Satis'], marker='o', color='blue')
ax[0].set_title('GÃ¼nlÃ¼k SatÄ±ÅŸ Trendi')

# --- 2. Grafik: SÃ¼tun GrafiÄŸi ---
ax[1].bar(veri['Kategori'], veri['Deger'], color='green')
ax[1].set_title('Kategori KarÅŸÄ±laÅŸtÄ±rmasÄ±')

# --- 3. Grafik: Histogram ---
ax[2].hist(veri['Dagilim'], bins=5, color='orange', edgecolor='black')
ax[2].set_title('Veri DaÄŸÄ±lÄ±mÄ± (Histogram)')

# GÃ¶rÃ¼nÃ¼mÃ¼ sÄ±kÄ±laÅŸtÄ±ralÄ±m ki baÅŸlÄ±klar iÃ§ iÃ§e girmesin
plt.tight_layout()
plt.show()
```

```python
## Grafik Yorumu (Histogramlar)
# - Grafik neyi netleÅŸtirdi?
# --Neden Bu GrafiÄŸi KullanÄ±rÄ±z? (Hepsi bir yerde analizler)
# -- plt.subplots(1, 3): Bu komutla Python'a "Bana yan yana 3 tane boÅŸ Ã§erÃ§eve ver" dedik.
# -- ax[0], ax[1], ax[2]: Bu indeksler sayesinde hangi grafiÄŸin hangi kutuya gideceÄŸini belirledik.
# -- figsize=(15, 5): GÃ¶rselin geniÅŸliÄŸini ve yÃ¼ksekliÄŸini ayarlayarak grafiklerin sÄ±kÄ±ÅŸmasÄ±nÄ± Ã¶nledik.
# -- Bir veri bilimci olarak projelerini sunarken, tek tek resimler gÃ¶stermek yerine bÃ¶yle derli toplu raporlar sunmak
# --- Ã§ok daha profesyonel durur.
```

```python
## Scatter Plot (SaÃ§Ä±lÄ±m GrafiÄŸi)
# - veri bilimcilerin dedektiflik aracÄ±dÄ±r.
# - Ä°ki farklÄ± sayÄ±sal deÄŸiÅŸken arasÄ±nda bir iliÅŸki (korelasyon) olup olmadÄ±ÄŸÄ±nÄ± anlamak iÃ§in kullanÄ±lÄ±r.
# - Ã–rneÄŸin; "Ders Ã§alÄ±ÅŸma sÃ¼resi arttÄ±kÃ§a sÄ±nav notu da artÄ±yor mu?" veya "Hava sÄ±caklÄ±ÄŸÄ± arttÄ±kÃ§a dondurma satÄ±ÅŸlarÄ± nasÄ±l deÄŸiÅŸiyor?"
# -- gibi sorularÄ±n cevabÄ±nÄ± bu noktalar kÃ¼mesinde buluruz.

# 1. Senaryo: Ã‡alÄ±ÅŸma SÃ¼resi vs. BaÅŸarÄ±
# -- Diyelim ki veri bilimi Ã§alÄ±ÅŸÄ±rken harcadÄ±ÄŸÄ±n saatler ile Ã§Ã¶zdÃ¼ÄŸÃ¼n Ã¶rnek sayÄ±sÄ± arasÄ±ndaki iliÅŸkiyi merak ediyorsun.
import pandas as pd
import matplotlib.pyplot as plt

# Veri setimizi oluÅŸturalÄ±m
veri = {
    'Calisma_Saati': [1, 2, 3, 3.5, 4, 5, 5.5, 6, 7, 8],
    'Cozulen_Ornek': [5, 12, 15, 14, 22, 28, 25, 35, 38, 45]
}

df_scatter = pd.DataFrame(veri)

# Scatter plot Ã§izimi (kind='scatter')
df_scatter.plot(x='Calisma_Saati', y='Cozulen_Ornek', kind='scatter',
                color='red', s=50, alpha=0.7)

plt.title('Ã‡alÄ±ÅŸma Saati ve Ã–rnek Ã‡Ã¶zÃ¼mÃ¼ Ä°liÅŸkisi')
plt.grid(True, linestyle='--', alpha=0.5) # Izgara eklemek okumayÄ± kolaylaÅŸtÄ±rÄ±r
plt.show()
```

```python
## Grafik Yorumu
# -Scatter Plot Okuma KÄ±lavuzu
# - Bu grafiÄŸe baktÄ±ÄŸÄ±nda noktalarÄ±n diziliÅŸine gÃ¶re 3 farklÄ± Ã§Ä±karÄ±m yapabilirsin:
# 1. Pozitif Ä°liÅŸki (YukarÄ± DoÄŸru): Noktalar sol alttan saÄŸ Ã¼ste doÄŸru bir hat oluÅŸturuyorsa, bir deÄŸiÅŸken arttÄ±kÃ§a diÄŸeri de artÄ±yor
# -- (YukarÄ±daki Ã¶rnekte olduÄŸu gibi). demektir.
# 2. Negatif Ä°liÅŸki (AÅŸaÄŸÄ± DoÄŸru): Noktalar sol Ã¼stten saÄŸ alta gidiyorsa, biri artarken diÄŸeri azalÄ±yor demektir.
# -- (Ã–rn: Ekran baÅŸÄ±nda geÃ§irilen vakit vs. uyku kalitesi).
# 3. Ä°liÅŸkisiz (DaÄŸÄ±nÄ±k): Noktalar rastgele daÄŸÄ±lmÄ±ÅŸsa, bu iki deÄŸiÅŸken arasÄ±nda bir baÄŸ yoktur diyebiliriz.
# Scatter plotlarda s (size) parametresi ile noktalarÄ±n bÃ¼yÃ¼klÃ¼ÄŸÃ¼nÃ¼, c (color) parametresi ile renklerini deÄŸiÅŸtirebilirsin.
# Hatta Ã¼Ã§Ã¼ncÃ¼ bir deÄŸiÅŸkeni (Ã¶rneÄŸin gÃ¼nÃ¼n yorgunluk seviyesi) noktalarÄ±n bÃ¼yÃ¼klÃ¼ÄŸÃ¼ olarak atayabilirsin. Buna da Bubble Chart (Balon GrafiÄŸi) denir!
```
